# Average Transaction Amount for Each Customer

```python
id=np.unique(my_data['customer_id'])
lst=[]
num=[]
tol=[]


for i in id:
    df1=my_data[my_data['customer_id'] ==i]
    num.append(len(df1))
    average_per_trans=sum(df1['amount'])/len(df1['amount'])
    lst.append(average_per_trans)
    tol.append(sum(df1['amount']))
    
    
```

```python
lst_A=[]
num_A=[]
tol_A=[]

for i in id:
    df1=df_August[df_August['customer_id'] ==i]
    num_A.append(len(df1))
    average_per_trans=sum(df1['amount'])/len(df1['amount'])
    lst_A.append(average_per_trans)
    tol_A.append(sum(df1['amount']))
```

```python
lst_S=[]
num_S=[]
tol_S=[]

for i in id:
    df1=df_September[df_September['customer_id'] ==i]
    num_S.append(len(df1))
    average_per_trans=sum(df1['amount'])/len(df1['amount'])
    lst_S.append(average_per_trans)
    tol_S.append(sum(df1['amount']))
```

```python
lst_O=[]
num_O=[]
tol_O=[]

for i in id:
    df1=df_October[df_October['customer_id'] ==i]
    num_O.append(len(df1))
    average_per_trans=sum(df1['amount'])/len(df1['amount'])
    lst_O.append(average_per_trans)
    tol_O.append(sum(df1['amount']))
```

```python
d = {'id': id, 'Average Transaction Amount': lst, 'Number of Transaction':num,'Total Transaction amount':tol,
     'Average Transaction Amount on August': lst_A, 'Number of Transaction on August':num_A,'Total Transaction amount on August':tol_A,
    'Average Transaction Amount on Sep': lst_S, 'Number of Transaction on Sep':num_S,'Total Transaction amount on Sep':tol_S,
    'Average Transaction Amount on October': lst_O, 'Number of Transaction on October':num_O,'Total Transaction amount on October':tol_O}
average_each_customers = pd.DataFrame(data=d)

average_each_customers
```

### Calculate the distance between customers and marchant

```python
from math import sin, asin, cos, radians, fabs, sqrt
 
EARTH_RADIUS=6371           # 6371km
 
def hav(theta):
    s = sin(theta / 2)
    return s * s
 
def get_distance_hav(lng0,lat0, lng1,lat1):
    lat0 = radians(lat0)
    lat1 = radians(lat1)
    lng0 = radians(lng0)
    lng1 = radians(lng1)
 
    dlng = fabs(lng0 - lng1)
    dlat = fabs(lat0 - lat1)
    h = hav(dlat) + cos(lat0) * cos(lat1) * hav(dlng)
    distance = 2 * EARTH_RADIUS * asin(sqrt(h))
    
    return distance
```

```python
id=np.unique(my_data['customer_id'])
sal=[]
for i in id:
    df=my_data[my_data['customer_id']==i]
    salary=sum(df[df["txn_description"] == "PAY/SALARY"]['amount'])
    sal.append(salary*4)

salaries['salary']=sal

gender_l=[]
for i in id:
    df=my_data[my_data['customer_id']==i]
    gender=df['gender'].unique()[0]
    if gender=='F':
        gen=1
    else:
        gen=0
    gender_l.append(gen)

salaries['gender']=gender_l
# salaries.head()

salaries=salaries.rename(columns={"balance": "mean_bal", "amount": "mean_amount"})
# salaries.head()

from statistics import mean 
from statistics import median
bal=[]
amo1=[]
amo2=[]
for customer_id in id:
    bal.append(median((my_data[my_data['customer_id']==customer_id])['balance']))
    amo1.append(max((my_data[my_data['customer_id']==customer_id])['amount']))
    amo2.append(min((my_data[my_data['customer_id']==customer_id])['amount']))

salaries['median_bal']=bal
salaries['max_amount']=amo1
salaries['min_amount']=amo2
salaries.head()
```
